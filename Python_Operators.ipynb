# Python Operators

## What are Operators in Python?

#### Operators are special symbols or keywords in Python used to perform operations on variables and values. Python supports a wide range of operators for different types of operations.





### 1. Arithmetic Operators

Used to perform mathematical operations

```python
a = 10
b = 3
print(a+b)
print(a-b)
print(a*b)
print(a/b)
print(a%b)
print(a//b)
print(a**b)

```

### 2. Comparison (Relational) Operators

Used to compare values. Returns True or False

```python
x = 20
y = 10
print(x==y)
print(x!=y)
print(x>y)
print(y<x)
print(x>=y)
print( y<=x)
```

### 3. Logical Operators

Used to combine conditional statements

```python
a = True
b = False
print(a and b, b and a)
print(a or b, b or a)
print( not a )
print( not b)
```

### 4. Assignment Operators

Used to assign values to variables

```python
x *= 2 #x = x*2
print(x)
```

```python
x %= 2
print(x)
```

```python
x/=2
print(x)
```

```python
x -= 2
print(x)
```

### 5. Bitwise Operators

Used to perform operations on bits

```python
a = 5 # 0101
b = 3 # 0011

print( a & b, a & b) # AND
print(a|b, a|b) #0R
print(a^b, a^b) #x0R
```

### 6. Membership Operators

Used to test if a sequence contains a value

```python
list1 = [1, 2, 3, 4]
print( 3 in list1)
```

```python
print(10 in list1)
```

```python
print(15 not in list1)
```

### 7. Identity Operators

Used to compare memory locations of two objects

```python
x = [1, 2, 3]
y = x  #refers to the same object as x
z = [1, 2, 3] # this is a different object

print(x is y)
```

```python
print( y is [1, 2, 3])
```

```python
print( x is z)
```

```python
print(x == z)
```

## Describe the difference between a function and a method

| Feature              | **Function**                                                              | **Method**                                                        |
| -------------------- | ------------------------------------------------------------------------- | ----------------------------------------------------------------- |
| **Definition**       | A **function** is a block of reusable code that performs a specific task. | A **method** is a function that is associated with an **object**. |
| **How it's used**    | Called on its own with parentheses                                        | Called **on an object** using dot `.` notation                    |
| **Example**          | `len([1, 2, 3])` or `print("Hello")`                                      | `"hello".upper()` or `[1, 2].append(3)`                           |
| **Belongs to**       | Global scope (not tied to a specific data type)                           | Belongs to a **class** (like `str`, `list`, `dict`)               |
| **Modifies object?** | Usually works with or returns new data                                    | Often **acts on** the object it's called on                       |
